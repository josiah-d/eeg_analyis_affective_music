```python
# imports
import matplotlib.pyplot as plt
import mne
```

```python
# make montage
biosemi16_montage = mne.channels.make_standard_montage('biosemi16')
montage = biosemi16_montage.plot()
```

```python
# add positional label
montage.text(0.5,0.975,'Anterior',horizontalalignment='center',
     verticalalignment='center', fontsize=14)
```

```python
# save figure
montage.savefig('../img/biosemi16_montage.jpg')
```
